```python
''' 
Code for Lottery game.. 
Combination of 3 numbers game. 
For each chance 1000 INR redcues. 
If all 3 numbers are EVEN, you get 300 INR as BONUS. 
If all 3 numbers are ODD, you get 300 INR as BONUS. 
If all 3 numbers are in sequence, you get 800 INR as BONUS. 
If all 3 numbers are same, you get 1000 INR as BONUS. 
'''

import random

# Initial Amount to begin the Game
InputAmt = 1000
game = 'PLAY'
print ("Welcome to the PlayWin game!!")
print ("You have INR", InputAmt, "to start the game \n")

class PlayWin:

    def randomNumber(self):
        # Randomly generated 3 numbers
          sysNum1 = random.randint(0,9)
          sysNum2 = random.randint(0,9)
          sysNum3 = random.randint(0,9)
          print("System Generated Numbers:",sysNum1, sysNum2, sysNum3)
          return sysNum1, sysNum2, sysNum3

    def oddNumber(self,InputAmt,sysNum1, sysNum2, sysNum3):
        if (sysNum1 % 2 == 0 ) and (sysNum2 % 2 == 0 ) and (sysNum3 % 2 == 0 ):
            InputAmt = InputAmt + 300
        return (InputAmt)

    def evenNumber(self,InputAmt, sysNum1, sysNum2, sysNum3):
        if (sysNum1 % 2 != 0 ) and (sysNum2 % 2 != 0 ) and (sysNum3 % 2 != 0 ):
            InputAmt = InputAmt + 300
        return (InputAmt)

    def seqNumber(self,InputAmt, sysNum1, sysNum2, sysNum3):
        if (((abs(sysNum1 - sysNum2)) == 2 and (abs(sysNum2 - sysNum3)) == 1 and (abs(sysNum3 - sysNum1)) == 1) or
            ((abs(sysNum1 - sysNum2)) == 1 and (abs(sysNum2 - sysNum3)) == 2 and (abs(sysNum3 - sysNum1)) == 1) or
            ((abs(sysNum1 - sysNum2)) == 1 and (abs(sysNum2 - sysNum3)) == 1 and (abs(sysNum3 - sysNum1)) == 2)):
            InputAmt = InputAmt + 800
        return InputAmt

    def sameNumber(self, InputAmt,sysNum1, sysNum2, sysNum3):
        if (sysNum1==sysNum3) and (sysNum2==sysNum3):
            InputAmt = InputAmt + 1000
        # InputAmt = int((InputAmt + 1000) if (sysNum1==sysNum3) and (sysNum2==sysNum3))
        return InputAmt

while (InputAmt > 99 and game != "STOP"):

    game = input("Do you wanna play the game? Enter 'Y' or 'N':")
    game = game.upper()
    
    if game == 'Y':
       InputAmt = InputAmt - 100
       game = 'PLAY'

       O=PlayWin()
       sysNum1, sysNum2, sysNum3 = O.randomNumber()
       InputAmt = O.sameNumber(InputAmt,sysNum1, sysNum2, sysNum3)
       InputAmt = O.seqNumber (InputAmt,sysNum1, sysNum2, sysNum3)
       InputAmt = O.evenNumber(InputAmt,sysNum1, sysNum2, sysNum3)
       InputAmt = O.oddNumber (InputAmt,sysNum1, sysNum2, sysNum3)
       
       print("Amount left is ",InputAmt,"\n")
    elif game == 'N':
       game = 'STOP'
       print("Thank you for Playing!! Final amount in your Kitty is INR", InputAmt)
    else:
       print("Invalid entry. Please try again!")
       game = input("Do you wanna play the game? Enter 'Y' or 'N':")
```
