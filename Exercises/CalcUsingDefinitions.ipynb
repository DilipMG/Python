```python
# Program to define simple calcualtor

#Define addition
def add(a, b):
    return a+b

#Define Subtraction
def sub(a, b):
    return a-b

#Define multiplication
def multiply(a, b):
    return a*b

#Define Division
def divide(a, b):
    return a/b

#Define Square
def Square(a,b):
    return a**b

#Define Modules
def Modules(a, b):
    return a%b

#Define AcceptInput
def AcceptInp():
    num1 = int(input("Enter the first number:"))
    num2 = int(input("Enter the seconds input number:"))
    return num1, num2

#Define Options
def Options():
    print("Welcome!! Please enter your choice \n")
    print("Enter 1 for Simple Calculator!")
    print("Enter 2 for Scientific Calculator!")
    Option=int(input("\nEnter your choice (1 or 2):"))
    return Option

#Define Simple Calc Options
def SimpleOptions():
    print("\nEnter 1 for Addition!")
    print("Enter 2 for Subtraction!")
    print("Enter 3 for Multiplcation!")
    print("Enter 4 for Division! \n")
    choice=int(input("Enter your choice (1/2/3/4):"))
    return choice


#Define Scientific Calc Options
def ScientificOptions():
    print("\nEnter 1 for Addition!")
    print("Enter 2 for Subtraction!")
    print("Enter 3 for Multiplcation!")
    print("Enter 4 for Division!")
    print("Enter 5 for Power function!")
    print("Enter 6 for Modules of Number! \n")
    choice=int(input("Enter your choice (1/2/3/4/5/6):"))
    return choice

#Define Results Print
def Results(choice, num1, num2):
    if choice == 1:
        print("\nSum of",num1,"&",num2,"is",add(num1, num2))
    elif choice == 2:
        print("\nSubraction of",num2,"from",num1,"is",sub(num1, num2))
    elif choice == 3:
        print("\nMultiplication of",num1,"&",num2,"is",multiply(num1, num2))
    elif choice == 4:
        print("\nDivision of",num1,"from",num2,"is",divide(num1, num2))
    elif choice == 5:
        print("\nPower of",num1,"to",num2,"is",Square(num1,num2))
    elif choice == 6:
        print("\nModules of",num1,"from",num2,"is",Modules(num1, num2))    
    else:
        print("\nInvalid choice! Please choose correct Choice again!")
```

```python
Option=Options()

if Option== 1:
    choice=SimpleOptions()
    num1,num2=AcceptInp()
    Results(choice, num1, num2)
elif Option ==2:
    choice=ScientificOptions()
    num1,num2=AcceptInp()
    Results(choice, num1, num2)
else:
    print("Invalid Options. Please try again")
```
