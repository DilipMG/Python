# Filtering a DataFrame

```python
#Filtering is abt Rows
Filter = stats.InternetUsers < 2
Filter
```

```python
stats[Filter]   # This will display all the rows with InternetUsers < 2. 
```

```python
#FilterBirth = stats.BirthRate > 40
#stats[FilterBirth]    

#The above two lines can be merged together as below. 

stats[stats.BirthRate > 40]
```

```python
# Combining multiple filters.. 'and' cannot be used since Python is not vectorized. hence '&' needs to used for condition.
stats[(stats.BirthRate > 40)  & (stats.InternetUsers < 2)]
```

```python
stats[stats.IncomeGroup == 'High income']
```

# Seaborn Package

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.rcParams['figure.figsize'] = 8,4
```

```python
stats.head()
```

```python
vis1 = sns.distplot(stats['InternetUsers'])
```

```python
vis1 = sns.distplot(stats['InternetUsers'],bins=30)
```

```python
vis2 = sns.boxplot(data=stats,x='IncomeGroup', y='BirthRate')
```

```python
vis3 = sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg = False)
```

```python
vis3 = sns.lmplot(data=stats,x='InternetUsers',y='BirthRate', \
                  fit_reg = False, hue = 'IncomeGroup', \
                 size = 5, aspect = 2)
```
